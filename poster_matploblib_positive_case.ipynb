```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("data_distribution.xlsx", index_col=0)
df
```

```python
plt.figure(figsize=(16,12))

# set width of bars
barWidth = 0.25

# set heights of bars
train_positive_rate = [d for d in df['train_positive_rate']]
val_positive_rate = [d for d in df['val_positive_rate']]
test_positive_rate = [d for d in df['test_positive_rate']]

# Set position of bar on X axis
r1 = np.arange(len(train_positive_rate))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

# set numbers on bars
positive_number = [[d for d in df['train_positive_number']], [d for d in df['val_positive_number']], [d for d in df['test_positive_number']]]
x = [[0, 1, 2, 3, 4, 5], [0.25, 1.25, 2.25, 3.25, 4.25, 5.25], [0.5, 1.5, 2.5, 3.5, 4.5, 5.5]]
y = [train_positive_rate, val_positive_rate, test_positive_rate]

for i in range(3):
    for j in range(6):
        plt.text(x[i][j], y[i][j]+0.01, positive_number[i][j], ha='center', size=23)

# Make the plot
plt.bar(r1, train_positive_rate, color='deepskyblue', width=barWidth, edgecolor='black', label='Train')
plt.bar(r2, val_positive_rate, color='azure', width=barWidth, edgecolor='black', label='Validation')
plt.bar(r3, test_positive_rate, color='steelblue', width=barWidth, edgecolor='black', label='Test')

# Add xticks on the middle of the group bars
plt.xlabel('Targets', size=30)
plt.ylabel('Positive Rate', size=30)
plt.xticks([r + barWidth for r in range(len(train_positive_rate))], ['safe to walk', 'car', 'scooter', 'bike', 'other'+'\n'+'obstacles', 'no obstacle'+'\n'+'in crosswalk'], fontsize=23)
plt.yticks([0.2, 0.4, 0.6], [0.2, 0.4, 0.6], fontsize=20)

plt.title("Positive Case Distribution", size=30)
plt.legend(fontsize=25)

# Create legend & Show graphic
plt.savefig("positive_case_distribution.png")
```
