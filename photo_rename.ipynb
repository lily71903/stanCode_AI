```python
import glob
import pandas as pd
import numpy
import matplotlib.pyplot as plt
from pandas.plotting import table
from PIL import Image, ExifTags
import os
```

```python
df = pd.read_excel("juneyo/juneyo_2889.xlsx", index_col=0)
df['path'] = [f for f in df['filename']]
df['filename'] = [f.split(".")[0] for f in df['filename']]
df
```

```python
new_filenames = ["{}.jpg".format(str(i).zfill(6)) for i in range(len(df))]
df.insert(1, "new_filename", new_filenames)
df = df.reset_index(drop=True)
df
```

```python
df.to_excel('processed_data/juneyo_2889.xlsx')
```

## Rename

```python
for i in range(len(df)):
    file_path = os.path.join("juneyo", df.iloc[i]["path"])
    new_path = os.path.join("processed_data/juneyo", df.iloc[i]["new_filename"])
    print("{} => {}".format(file_path, new_path))
    
    try:
        img = Image.open(file_path)
        img.save(new_path)
    
    except Exception as err:
        print("cannot handle {}".format(file_path))
        print(err)
```
